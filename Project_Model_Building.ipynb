## 4. Feature Engineering

### 4.1 Drop Unwanted Columns

```python
df.head(4)
```

```python
df.drop(columns=['Property_Name', 'Location','Availability','Bathroom'], inplace =  True)
print('Shape of data :', df.shape)
```

### 4.2 Label Encoding for Categorical Columns

```python
le = LabelEncoder()
```

```python
for column in df.describe(include='object').columns:
    df[column] = le.fit_transform(df[column])
```

## 5. Model Building

```python
df.head()
```

### 5.1 Train Test Split

```python
X =  df.drop('Price_Lakh', axis = 1)
y =  df['Price_Lakh']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state = 12)
```

```python
print(X_train.shape, X_test.shape)
```

### 5.2 Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
linear.fit(X_train, y_train)

print("Training Accuracy = ", linear.score(X_train, y_train))
print("Test Accuracy     = ", linear.score(X_test, y_test))
```

### 5.3 xgboost

```python
import xgboost as xgb

# Initialize and train XGBoost
xg_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=100,
    random_state=12
)
xg_model.fit(X_train, y_train)

# Print scores (R²) in your exact format
print("Training Accuracy = ", xg_model.score(X_train, y_train))
print("Test Accuracy     = ", xg_model.score(X_test, y_test))
```

### 5.4 Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 1000, max_depth=5, random_state = 12)
rf.fit(X_train, y_train);

print("Training Accuracy = ", rf.score(X_train, y_train))
print("Test Accuracy     = ", rf.score(X_test, y_test))
```

## 6. Final Model Evaluation

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    accuracy = model.score(test_features, test_labels)

    print('Average Error  = {:0.4f} degrees'.format(np.mean(errors)))
    print('Model Accuracy = {:0.4f} %'.format(accuracy))
```

```python
evaluate(xg_model, X_train, y_train)
```

```python
evaluate(xg_model, X_test, y_test)
```
